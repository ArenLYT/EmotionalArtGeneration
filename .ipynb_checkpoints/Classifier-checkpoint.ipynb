```python
from fastai.data.all import *
from fastai.vision.all import *
import fastai
import pandas as pd
import statistics
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('ArtEmisv1.csv')
df
```

```python
def modo(x):
    if  x.value_counts()[0]>=sum(x.value_counts())*0.8 :
        return pd.Series.mode(x)
    else:
        return x.value_counts()[:3].index.tolist()
```

```python
dfemo=df

dfemo=dfemo.groupby(["art_style","painting"])["emotion"].agg(modo).reset_index()

dfemo
```

```python
t=type(dfemo.emotion[0])
dfemo=dfemo[dfemo["emotion"].apply(lambda x: type(x) !=t )].reset_index()
dfemo
```

```python
dfemo = dfemo.drop('index', axis=1)
dfemo
```

```python
dfemo['path']= dfemo.apply(lambda x: 'dataset\\wikiart\\'+x['art_style']+"\\"+ x['painting']+".jpg", axis=1)
dfemo
```

```python
for i in dfemo.path:
    try:
        im=Image.open(i)
    except:
        dfemo=dfemo.drop(dfemo[dfemo.path==i].index)
dfemo
```

```python
imageDB=DataBlock(blocks=(ImageBlock,CategoryBlock),
                  get_x=ColReader('path'),
                  get_y=ColReader('emotion'),
                  item_tfms=Resize(224,ResizeMethod.Pad, PadMode.Zeros))
```

```python
dls = imageDB.dataloaders(dfemo,bs=16)
dls.show_batch()
```

```python
opt_func = partial(ranger)
model=xresnet34_deeper
learn = vision_learner(dls, model,pretrained=False, metrics=error_rate,opt_func=opt_func)
```

```python
lr=learn.lr_find(suggest_funcs=(valley))
learn.recorder.plot_lr_find()
```

```python
learn.fit_flat_cos(30,lr,pct_start=0.75, cbs=[SaveModelCallback (monitor='valid_loss', 
                                                                  comp=None, 
                                                                  min_delta=0.0,
                                                                  fname='classification_model_xresnet34_deeper_refined', 
                                                                  every_epoch=False,
                                                                  at_end=False,
                                                                  with_opt=False, 
                                                                  reset_on_fit=True)])
```

```python
model=learn.load('classification_model_xresnet34_deeper_refined')
```
