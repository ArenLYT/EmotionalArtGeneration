```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
from PIL import Image, ImageStat
import glob
```

```python
df=pd.read_csv('ArtEmisv2.csv')
df
```

```python
df.info()
```

```python
processed=df.groupby(["art_style","painting"])["emotion"].apply(statistics.mode).reset_index()
```

```python
processed
```

```python
print(processed.emotion.unique(),len(processed.emotion.unique()))
```

```python
plt.figure(figsize=(10,10))
plt.hist(processed.emotion,bins=range(0,9), rwidth=0.8,align="left")
plt.title('Histogram of Classes')
```

```python
print(processed.art_style.unique(),len(processed.art_style.unique()))
```

```python
plt.figure(figsize=(100,20))
plt.hist(processed.art_style,bins=range(0,28), rwidth=0.8,align="left")
plt.title('Histogram of Art Style')
```
